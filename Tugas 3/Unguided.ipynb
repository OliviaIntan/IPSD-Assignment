# Cek Outlier

```python
numerical_train = train_features.select_dtypes(include=['float64', 'int64'])
numerical_test = test_features.select_dtypes(include=['float64', 'int64'])

def detect_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    return (data < (Q1 - 1.5 * (Q3 - Q1))) | (data > (Q3 + 1.5 * (Q3 - Q1)))

for column in numerical_train.columns.intersection(numerical_test.columns):
    train_outliers = detect_outliers(numerical_train[column]).sum()
    test_outliers = detect_outliers(numerical_test[column]).sum()

    print(f"Fitur: {column} | Outlier Train: {train_outliers} | Outlier Test: {test_outliers}")

    plt.figure(figsize=(10, 5))
    sns.boxplot(data=[numerical_train[column], numerical_test[column]], palette='Set3')
    plt.xticks([0, 1], ['Train', 'Test'])
    plt.title(f'Outlier Comparison: {column}')
    plt.ylabel(column)
    plt.show()
```

#Visualisasi

```python
test_features.hist(bins=20, figsize=(15, 10))
plt.suptitle('Distribusi Data Test', fontsize=16)
plt.show()

train_features.hist(bins=20, figsize=(15, 10))
plt.suptitle('Distribusi Data Train', fontsize=16)
plt.show()
```

```python
fitur_x = 'pendapatan'
fitur_y = 'belanja_daging'

# Scatter plot untuk Data Train
plt.figure(figsize=(12, 6))
sns.scatterplot(x=train_features[fitur_x], y=train_features[fitur_y], color='pink', alpha=0.6, label='Data Train')
plt.title('Scatter Plot Data Train', fontsize=16)
plt.xlabel(fitur_x)
plt.ylabel(fitur_y)
plt.legend()
plt.grid(True)
plt.show()

# Scatter plot untuk Data Test
plt.figure(figsize=(12, 6))
sns.scatterplot(x=test_features[fitur_x], y=test_features[fitur_y], color='green', alpha=0.6, label='Data Test')
plt.title('Scatter Plot Data Test', fontsize=16)
plt.xlabel(fitur_x)
plt.ylabel(fitur_y)
plt.legend()
plt.grid(True)
plt.show()
```
